```python
import requests
from io import BytesIO, StringIO
from zipfile import ZipFile

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00312/dow_jones_index.zip"

with ZipFile(BytesIO(requests.get(url).content), "r") as zf:
    print(zf.namelist())
```

```python
# The mode parameter should be 'r' to read an existing file, 'w' to truncate and write a new file, 
# 'a' to append to an existing file, or 'x' to exclusively create and write a new file. 
# If mode is 'x' and file refers to an existing file, a FileExistsError will be raised. 
# If mode is 'a' and file refers to an existing ZIP file, then additional files are added to it. 
# If file does not refer to a ZIP file, then a new ZIP archive is appended to the file. 
# This is meant for adding a ZIP archive to another file (such as python.exe). 
# If mode is 'a' and the file does not exist at all, it is created. If mode is 'r' or 'a', the file should be seekable.

with ZipFile(BytesIO(requests.get(url).content), "r") as zf:
    for archivo in ["dow_jones_index.data"]:
        try:
            data = zf.read(archivo)
        except KeyError:
            print('No se encontró el archivo dentro del archivo zip'.format(archivo))
        else:
            print(archivo, ':')
            print(data)
        print()
```

```python
type(data)
```

```python
data_nuevo = str(data,'utf-8')
data_nuevo = StringIO(data_nuevo) 
data_nuevo = pd.read_csv(data_nuevo)
display(data_nuevo)
```

```python
data_nuevo.info()
```

```python
def remover_simbolo_precio(precios):
    import pandas as pd
    Transformacion = [PR.split('$')[1] for PR in precios]
    Transformacion = pd.DataFrame(Transformacion).astype(float)
    return Transformacion
```

```python
data_nuevo["open"] = remover_simbolo_precio(data_nuevo["open"])
data_nuevo["high"] = remover_simbolo_precio(data_nuevo["high"])
data_nuevo["low"] = remover_simbolo_precio(data_nuevo["low"])
data_nuevo["close"] = remover_simbolo_precio(data_nuevo["close"])
data_nuevo.info()
```
